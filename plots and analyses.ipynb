```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

load dataset

```python
data = pd.read_csv("2020-08-24-listings.csv",low_memory=False)
data.head()
```

describe dataset info

```python
data.describe(include='object').T
```

# 1) Which neighbourhood is preferable for renting based on guest review scores?

```python
use_cols = ['neighbourhood','review_scores_accuracy', 'review_scores_cleanliness', 'review_scores_checkin',
               'review_scores_communication', 'review_scores_location', 'review_scores_value']

Nei = data[use_cols].groupby('neighbourhood').mean().dropna()
```

```python
for col in Nei.columns:
    Nei[col].sort_values().head(20).plot.barh(figsize=(10, 10));
    sns.despine()
    plt.title(f'Top 20 mean of {col} by neighbourhood');
    plt.show()
```

Neighbourhoods in the center of London are more attractive by location for guests, especially on the north side of the Thames (average review score by location):
+ Kensington and Chelsea (~9.76)
+ Westminster (~9.73)
+ Camden (~9.73)

The best regions in general (corresponding to overall mean guest review scores) are located on the Southwest of London (overall average review score):
+ Kingston Upon Thames (~9.70)
+ Richmond Upon Thames (~9.69)

# 2) How host/listings properties are related with guest review scores?

```python
use_cols = ['room_type', 'instant_bookable', 'host_is_superhost','review_scores_accuracy',
            'review_scores_cleanliness', 'review_scores_checkin',
            'review_scores_communication', 'review_scores_location', 'review_scores_value']
use_data = data[use_cols]
class_cols = ['room_type', 'instant_bookable', 'host_is_superhost']
```

```python
RT = use_data.groupby('room_type').mean().dropna()
for col in RT.columns:
    RT[col].sort_values().plot.bar(figsize=(10, 10));
    plt.title(f'Top 20 mean of {col} by room type');
    plt.show()
```

```python
IB = use_data.groupby('instant_bookable').mean().dropna()
for col in IB.columns:
    IB[col].sort_values().plot.bar(figsize=(10, 10));
    plt.title(f'Top 20 mean of {col} by instant bookable');
    plt.show()
```

```python
HIS = use_data.groupby('host_is_superhost').mean().dropna()
for col in HIS.columns:
    HIS[col].sort_values().plot.bar(figsize=(10, 10));
    plt.title(f'Top 20 mean of {col} by host is superhost');
    plt.show()
```

+ Guest reviews for private and shared rooms are consistently greater than for other room types (except for Location)
+ For Location review scores, the best room types are entire home/apartments and hotel rooms
+ If additional approval from host is not required for booking (instant bookable listings), then guest reviews are consistently better (except for Location)
+ If host has the status of Superhost, then guest reviews are consistently better (except for Location)

# 4 How much are hosts making from renting to tourists (compare that to long-term rentals)

```python
use_cols = ["host_name","price"]
use_data = data[use_cols]
```

```python
use_data.head()
```

```python
use_data["price"] = use_data["price"].apply(lambda x: float(str(x).replace("$",'').replace(",",'')))
use_data.groupby("host_name").sum()["price"].sort_values(ascending=False).head(10).plot.barh(figsize=(10, 10));
plt.title(f'Top 10 rich hosts');
plt.show()
```
